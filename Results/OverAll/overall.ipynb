```python
import pandas as pd 
import numpy as np
import os      
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 #  Server process time 0.24, 0.50, 1.02, 1.62,  3.33 
Server_Process_time=[ 0.24, 0.50, 1.02, 1.62, 3.33]
```

# Max temperature

```python
RR_max_temp=[ 69.2, 71.5, 75.4, 77.7, 76.8]
CPUbased_temp=[ 70.0, 72.4, 73.8, 77, 76.8]
Prop_temp=[63.4, 68.5, 73, 74, 76.5]
```

```python
sns.set()
plt.plot(Server_Process_time, RR_max_temp, label="RR")
plt.plot(Server_Process_time, CPUbased_temp, label= "CPU temperature based")
plt.plot(Server_Process_time, Prop_temp, label="Proposed ")
plt.xlabel("Average process time (s)")
plt.ylabel("Max CPU temperature (deg C)")
plt.legend()
```

# Max local fan rotation

```python
RR_max_fan=[ 60, 71.7, 86.6, 95.7, 82.7]
CPUbased_max_fan=[ 60, 62.7, 87.8, 85.9, 91]
Prop_max_fan=[40.78, 55.7, 73.7, 80.8, 84.7]
```

```python
sns.set()
plt.plot(Server_Process_time, RR_max_fan, label="RR")
plt.plot(Server_Process_time, CPUbased_max_fan, label="CPU temperature based")
plt.plot(Server_Process_time, Prop_max_fan, label="Proposed")
plt.xlabel("Average process time (s)")
plt.ylabel("Max local fan rotation speed (%)")
plt.legend()
```
